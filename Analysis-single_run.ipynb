```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
M = 1      #Number of repetitions
N = 3000    #Number of iterations
batch_size = 32
model_out = './results/spiking_bp_M_%d_N_%d_batchsize_%d.pkl'%(M, N, batch_size)
results_out = './results/spiking_bp_M_%d_N_%d_batchsize_%d_results.pkl'%(M, N, batch_size)
with open(results_out,"rb") as f:
    to_save=pickle.load(f,encoding='latin1')
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
```

```python
# np.cumsum(to_save['accs'][0],dtype=float)[:-10].shape
```

```python
plt.plot(to_save['accs'][0])
plt.plot(moving_average(to_save['accs'][0],50),'o')
plt.title("Accuracy")
plt.xlabel("Iteration")
plt.ylabel("%Accuracy")
plt.show()

plt.plot(to_save['losses'][0])
plt.plot(moving_average(to_save['losses'][0],50),'o')
plt.title("Loss")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.show()
```
