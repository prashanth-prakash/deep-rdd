```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
M = 1      #Number of repetitions
N = 2000    #Number of iterations
batch_size = 32

model_out = './results/run2_valeta_valsigma_spiking_bp_M_%d_N_%d_batchsize_%d.pkl'%(M, N, batch_size)
results_out = './results/run2_valeta_valsigma_spiking_bp_M_%d_N_%d_batchsize_%d_results.pkl'%(M, N, batch_size)
with open(results_out,"rb") as f:
    to_save=pickle.load(f,encoding='latin1')
```

```python
eta_dict = dict()
sigma_dict = dict()

for iter_count in range(2000):
    for count_sigma in range(6):
        for count_eta in range(6):
            to_save['accs'][0][0:2000][iter_count][count_sigma,count_eta]
                
```

```python
val_eta= np.logspace(-1,0,6)
val_sigma = np.logspace(-1,2,6)
print(val_eta,val_sigma)
```

```python
print(to_save['accs'][0][-1])
plt.figure(figsize = (10,15))
plt.imshow(to_save['accs'][0][-1])
plt.xticks(np.arange(6),np.round(val_sigma,2))
plt.yticks(np.arange(6),val_eta)
plt.xlabel("Sigma")
plt.ylabel("Learning rate")
plt.colorbar()
plt.title("Accuracy after 2000 iterations")
plt.show()
```

```python
plt.figure(figsize = (10,15))
plt.imshow(to_save['losses'][0][-1])
plt.xticks(np.arange(6),np.round(val_sigma,2))
plt.yticks(np.arange(6),val_eta)
plt.xlabel("Sigma")
plt.ylabel("Learning rate")
plt.colorbar()
plt.title("Losses after 2000 iterations")
```
