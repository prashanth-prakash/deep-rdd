```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
M = 1      #Number of repetitions
N = 2000    #Number of iterations
batch_size = 32

model_out = 'spiking_bp_M_%d_N_%d_batchsize_%d-Copy1.pkl'%(M, N, batch_size)
results_out = 'spiking_bp_M_%d_N_%d_batchsize_%d_results-Copy1.pkl'%(M, N, batch_size)
with open(results_out,"rb") as f:
    to_save=pickle.load(f)
```

```python
val_eta= np.logspace(-1,0,6)
print(to_save)
```

```python
for i in range(6):
    plt.plot(to_save['accs'][:,:,i][0],label=val_eta[i])
plt.legend(loc='best')
# plt.title("Accuracy")
plt.xlabel("Iterations")
plt.ylabel("Accuracy")
plt.show()
for i in range(6):
    plt.plot(to_save['losses'][:,:,i][0],label = val_eta[i])
plt.legend(loc='best')
# plt.title("Accuracy")
plt.xlabel("Iterations")
plt.ylabel("Losses")
plt.show()
```
